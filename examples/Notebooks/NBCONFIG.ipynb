# Configuration for notebook style

This notebook is intended to be run by other notebooks through the **%run** ipython magic.  The running notebook will then be styled nicely, have the most common pyparty imports, have pre-set figure size and plotting backend (svg), and will suppress logging in certain cases.

#### Set default plotting fig size and backend to SVG for nicer plots if rendering as .pdf.  If user doesn't load notebook with pylab, images will not always render, so I actually am going to enfore the pylab call indirectly by making a call to **plt**.

```python
class ConfigError(Exception):
    """ """
```

```python
try:             
    plt.rcParams['figure.figsize'] = 6, 4
except NameError:
    raise ConfigError("matplotlib plt not imported; please rerun notebook with flag"
                      "--pylab=inline.  Configuration requires pylab enabled.")
%config InlineBackend.figure_format = 'svg'
```
